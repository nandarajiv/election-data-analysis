##### i. Regionalism, Assembly Elections and Voter Turnout

There are a few major transformations in Indian politics that has created circumstances that would come into play in the 2024 election. <br/>
If the 2014 elections marked the start of a more nationalist and pro-India narrative, since then marked the rise of regionalism, which has significantly grown, and is set to potentially be a major influence in the upcoming 2024 election. ,<br/>
There have been National parties looking to tap into the regional ethos, especially recenelty with new coalitions coming together. <br/><br/>

In light of this, an important thing to track is the rise of the regionalism ethos since 2014. <br/>
A measure for this is in the percentage turnout of voters in respective assembly elections versus state elections.,<br/>
Apart from the Lok Dhaba dataset provided, I downloaded a dataset for all assembly elections for respective states, selected a time period post 2013, and looked the turnout of voters by state for assmebly elections. <br/>
From the dataset provided, similar data of turnout of voters by state for general elections was also obtained, and contrasted.

```python
# basic imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load GE data into dataframe
df_ge = pd.read_csv('All_States_GE.csv')
df_ge = df_ge.loc[df_ge['Year'] >= 2014]
df_ge
```

```python
# get the turnout per state for 2014 onwards
cand_1_position_ge = df_ge.loc[df_ge['Position'] == 1]

cand_1_position_ge
```

```python
# load AE data into dataframe
df_ae = pd.read_csv('All_States_AE.csv')
df_ae = df_ae.loc[df_ae['Year'] >= 2013]
df_ae
```

```python
cand_1_position_ae = df_ae.loc[df_ae['Position'] == 1]

cand_1_position_ae
```

```python
grouped_df_ae = cand_1_position_ae.groupby(['State_Name'])
av_turnout_ae = grouped_df_ae['Turnout_Percentage'].mean()
print(av_turnout_ae)
```

```python
grouped_df_ge = cand_1_position_ge.groupby(['State_Name'])
av_turnout_ge = grouped_df_ge['Turnout_Percentage'].mean()
av_turnout_ge
```

```python
mean_ge = grouped_df_ge['Turnout_Percentage'].mean()
mean_ae = grouped_df_ae['Turnout_Percentage'].mean()

plt.bar(av_turnout_ge.index, av_turnout_ge.values, alpha=0.5, color='r')
plt.bar(av_turnout_ae.index, av_turnout_ae.values, alpha = 0.5, color='y')

plt.title('Mean Turnout Percentage by State')
plt.legend(['General Election', 'Assembly Election'], fontsize=7, loc='upper left')

plt.xticks(rotation=90)

plt.show()
```

Nearly all states demonstrate greater turnout rates for assembly elections rather than general elections. This is especially true in states like states and UTs like Nagaland, Manipur, Tripura, Meghalaya, J&K, etc., while significantly less observable in West Bengal, Kerala, Tamil Nadu, UP, Bihar, etc. 

This could point to some sense of lack of representation or weight to one's vote in a General Election when compared to a state election. In any case, this is a potential avenue for national parties to look into, to ally with regional parties that perform in assmebly election in these states and UTs, which has been a strategy that is falling into place for this election in some cases. <br/><br/>


#### ii. Declared Education and Success in Elections

The dataset provides a column that contains details of a candidate's educational background, such as whether a candidate has a doctorate, post gradaute, 10th pass, 8th pass,or whether they are a graduate professional, or illiterate, etc. <br/>
This datapoint could be weighed against candidates' success, to find out what appeals to voters. <br/>

To expand the scope of this exercise beyind just a candidate winning or not winning, since win margins may be very low among top candidates, and losing by a small margin is also some measure of success, the factor considered is the <b>security deposit</b>.

A candidate recieved their security deposit back only if they have gotten greater than 1/6th of the valid votes in their constituency for a general election in India, and this is the measure used here for a successful candidature.

```python
df = pd.read_csv('All_States_GE.csv')

condition = df['Year'] != 2019
df = df.drop(df[condition].index)

df
```

```python
def get_ed_score(x):
    if x == "Illiterate":
        return 0
    elif x == "8th Pass" or x == "10th Pass" or x == "12th Pass":
        return 1
    elif x == "Graduate" or x == "Graduate Professional":
        return 2
    elif x == "Post Graduate" or x == "Doctorate":
        return 3
    else:
        return 1.5
    
df['ed_cat'] = df['MyNeta_education'].apply(get_ed_score)

df['ed_cat']
```

```python
no_deposit_lost_df = df.loc[df['Deposit_Lost'] == 'no']

no_deposit_lost_df
```

```python
print(df['ed_cat'].mean())
print(no_deposit_lost_df['ed_cat'].mean())

plt.bar(x = "8th/10th/12th Pass", height = 1, color = 'c')
plt.bar(x = "Average Candidate", height = df['ed_cat'].mean(), color = 'r')
plt.bar(x = "Graduate and/or Professional", height = 2, color = 'c')
plt.bar(x = "Successful Candidate", height = no_deposit_lost_df['ed_cat'].mean(), color = 'g')
plt.bar(x = "PG or doctorate", height = 3, color = 'c')



plt.xticks(rotation = 69)
plt.title("Education and Candidate Success")
plt.show()
```

The average education for candidates that do not lose deposit (aka, they get more than one sixth of the total number of valid votes in their constituency) is significantly above the average education of all candidates.
This could mean either or both of the following:
- educated candidates get voted in more 
- educated candidates are likely tob evaluate their chances of getting a significant chunk of the votes more effectively, hence risking their deposit less

In any case, there is correlation between education and performance of a candidate. <br/><br/>

### iii. Win Margin and Turnout Percentage

Trying to establish a relationship between win margin and turnout percentage.

```python
df = pd.read_csv('All_States_GE.csv')
condition = df['Year'] != 2019
df = df.drop(df[condition].index)

cand_1_position = df.loc[df['Position'] == 1]
cand_1_position
```

```python
plt.scatter(df['Turnout_Percentage'], df['Margin_Percentage'], alpha = 0.2)

plt.ylabel("Margin Percentage of Victory")
plt.xlabel("Turnout Percentage")
plt.title("Margin Percentage of Victory vs Turnout Percentage")

plt.show()
```

Clearly, they are independent factors, and there is significant range of margin percentage, despite variation in turnout.<br/><br/>
